***1. Create a Circle class and initialize it with radius. Make two methods getArea and getCircumference inside this class.***

```
class Circle:
  def __init__(self, radius):
    self.radius = radius
    
  def getArea(self):
    area = 3.14 * self.radius * self.radius
    print("Area:", area)
    
  def getCircumference(self):
    c = 2 * 3.14 * self.radius
    print("Cicrumference:",c)
  
c1 =Circle(int(input("Enter radius:")))
c1.getArea()
c1.getCircumference()
'''
Output:
Enter radius:10
Area: 314.0
Cicrumference: 62.800000000000004
'''
```

***2. Create a Temperature class. Make two methods:***
1. convertFahrenheit - It will take celsius and will print it into Fahrenheit.
2. convertCelsius - It will take Fahrenheit and will convert it into Celsius.

```
class temp:
  def __init__(s,t ):
    s.t =t
  
  def CtoF(s):
    f= ((s.t * 9 ) / 5 ) +32
    print("Celsius to Farenheit: ", f , "F")
    
  def FtoC(s):
    c= (( s.t - 32 ) * 5) /9
    print("Farenheit to Celsius: ", c , "C")
    
t1 =temp(int(input("Enter temperature in Celsius:")))
t1.CtoF()

t2 =temp(int(input("\nEnter temperature in Farenheit:")))
t2.FtoC()

'''
Output:
Enter temperature in Celsius:10
Celsius to Farenheit:  50.0 F

Enter temperature in Farenheit:40
Farenheit to Celsius:  4.444444444444445 C
'''
```
