**Types of Arguments:**
1. ** Default :**
```
def greet(name, msg = "Good morning!"):
      print("Hello",name + ', ' + msg)
      
      print("Bob")   #bob is passed for name, msg is by default
```

2. **Keyword**
```
greet(name = "Alice",msg = "How do you do?")
```
Args can be out of order.

3.** Arbitary**: 

 No. of args is not known at first.

```
def greet(*names):
   """This function greets all
   the person in the names tuple."""

   # names is a tuple with arguments
   for name in names:
       print("Hello",name)

greet("Monica","Luke","Steve","John")

```
**Anonymous Function:**
The function does not have a name. keyword *lambda* is used instead of *def*.


